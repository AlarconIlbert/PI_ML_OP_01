```python
def count_unique_values_num(data, column_name):
    # Obtener los valores únicos y su recuento
    unique_values = data[column_name].value_counts().reset_index()
    unique_values.columns = [column_name, 'Count']
    unique_values = unique_values.sort_values(by=column_name)

    return unique_values
```

```python
column_name = 'namescast'
```

```python
df = df.replace("\[", '', regex=True)
df = df.replace("\]", '', regex=True)
df = df.replace("'", '', regex=True)
df = df.replace('"', '', regex=True)
```

```python
def count_unique_values_text(data, column_name):
    # Obtener los valores únicos y su recuento
    unique_values = data[column_name].value_counts().reset_index()
    unique_values.columns = [column_name, 'Count']
    unique_values = unique_values.sort_values(by='Count', ascending=False)

    return unique_values
```

```python
result = count_unique_values_text(df, column_name)
```
